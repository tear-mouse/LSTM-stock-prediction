```python
import torch.nn as nn # type: ignore
import torch.optim as optim # type: ignore
import pandas as pd # type: ignore
import numpy  as np # type: ignore
import torch # type: ignore
from torch.optim import Adam # type: ignore
from torch.utils.data import DataLoader # type: ignore
import matplotlib.pyplot as plt # type: ignore
from sklearn.model_selection import train_test_split # type: ignore
from sklearn.preprocessing import MinMaxScaler # type: ignore
from torch.optim.lr_scheduler import StepLR # type: ignore
from torch.utils.tensorboard import SummaryWriter
import os
```

```python
#数据预处理
stock_data = pd.read_csv('history_Index_k_day_data.csv', encoding='utf-8')
stock_data['date'] = pd.to_datetime(stock_data['date'], format='%Y-%m-%d')
stock_data

print(stock_data.head())  # 查看前几行数据
```

```python
# 训练相关超参数
save_dir = './models'  # 定义模型保存目录
os.makedirs(save_dir, exist_ok=True)  # 创建保存目录

config = {
    "batch_size": 32,            # 批量大小
    "learning_rate": 0.01,       # 学习率
    "epochs": 100,                # 训练轮数
    "step_size": 10,              # 学习率调整步长
    "gamma": 0.2,                 # 学习率衰减因子
    "window_size": 90,            # 时间窗口大小 (过去多少天的数据作为输入)
    "early_stop": 5,            # 早停步数
    "device": 'cuda' if torch.cuda.is_available() else 'cpu',  # 训练设备
    "save_path": os.path.join(save_dir, 'best_model.pth')  # 生成路径
}

```

```python
# 定义LSTM模型
class LSTM(nn.Module):
    def __init__(self, input_size=2, hidden_size=32, output_size=1, num_layers=2, dropout_rate=0.2):
        super(LSTM, self).__init__()

        # LSTM 网络结构
        self.input_size = input_size       # 输入特征维度，如 (close, volume)
        self.hidden_size = hidden_size     # LSTM 隐藏层神经元数量
        self.output_size = output_size     # 输出维度 (预测的股价)
        self.num_layers = num_layers       # LSTM 层数
        self.dropout_rate = dropout_rate   # 随机失活，防止过拟合
        self.num_directions = 2            # 单向 LSTM，可改为 2 代表双向 LSTM
         #LSTM层 输入维度为input_size，隐藏状态维度为hidden_size，batch_first=True
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout_rate)
        #全连接层 输入维度为hidden_size，输出维度为output_size  
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        #初始化隐藏状态和细胞状态
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        #LSTM层 输入x，隐藏状态h0，细胞状态c0       
        out, _ = self.lstm(x, (h0, c0))
        #全连接层 输入out，输出out[:, -1, :]
        out = self.fc(out[:, -1, :])
        return out


dataset = stock_data[['close', 'volume']].values

# 数据处理
def create_dataset(dataset, window_size):
    X, Y = [], []
    for i in range(len(dataset) - window_size):
        X.append(dataset[i:i + window_size])
        Y.append(dataset[i + window_size, 0])  # 预测下一天的股价
    return np.array(X), np.array(Y)

# 创建数据集
X, Y = create_dataset(dataset, config['window_size'])

# 训练集:验证集:测试集 = 6:2:2
train_size = int(0.6 * len(dataset))
val_size = int(0.2 * len(dataset))
test_size = len(dataset) - train_size - val_size



scaler = MinMaxScaler()
# 假设你的输入特征数据是 X，形状为 (n_samples, 2)


# 划分数据
X_train = X[:train_size]
Y_train = Y[:train_size]

X_test = X[train_size+val_size:]
Y_test = Y[train_size+val_size:]

X_val = X[train_size:train_size+val_size]
Y_val = Y[train_size:train_size+val_size]

# 对每个时间步的特征进行归一化
scaler = MinMaxScaler()
for i in range(len(X_train)):
    X_train[i] = scaler.fit_transform(X_train[i])
for i in range(len(X_val)):
    X_val[i] = scaler.transform(X_val[i])
for i in range(len(X_test)):
    X_test[i] = scaler.transform(X_test[i])

# 对Y进行归一化
scaler_y = MinMaxScaler()
Y_train = scaler_y.fit_transform(Y_train.reshape(-1, 1))
Y_val = scaler_y.transform(Y_val.reshape(-1, 1))
Y_test = scaler_y.transform(Y_test.reshape(-1, 1))


# 转换为PyTorch张量
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
Y_train_tensor = torch.tensor(Y_train, dtype=torch.float32)
Y_test_tensor = torch.tensor(Y_test, dtype=torch.float32)
Y_val_tensor = torch.tensor(Y_val, dtype=torch.float32)


```

```python
# 创建三个数据加载器
train_loader = DataLoader(
    dataset=torch.utils.data.TensorDataset(X_train_tensor, Y_train_tensor),
    batch_size=32,
    shuffle=False
)

val_loader = DataLoader(
    dataset=torch.utils.data.TensorDataset(X_val_tensor, Y_val_tensor),
    batch_size=32,
    shuffle=False  # 验证时不需要打乱
)

test_loader = DataLoader(
    dataset=torch.utils.data.TensorDataset(X_test_tensor, Y_test_tensor),
    batch_size=32,
    shuffle=False  # 测试时不需要打乱
)
# 检查设备是否可用，若可用则使用GPU，否则使用CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
import math
from tqdm import tqdm

def trainer(train_loader, val_loader, model, config, device):
    """
    训练 LSTM 模型，结合早停、模型保存、动态学习率调整等技术。
    
    参数：
    - train_loader: 训练数据的 DataLoader
    - val_loader: 验证数据的 DataLoader
    - model: 要训练的 LSTM 模型
    - config: 训练参数配置（字典格式）
    - device: 设备（'cuda' or 'cpu'）
    """
    
    # 选择损失函数（L1Loss = MAE，MSELoss = MSE）
    criterion = nn.MSELoss(reduction='mean')  # 使用均方误差（MSE）作为损失函数
    optimizer = optim.Adam(model.parameters(), lr=config['learning_rate'])  # Adam 优化器
    scheduler = StepLR(optimizer, step_size=config['step_size'], gamma=config['gamma']) # 每 10 轮学习率降低 

    writer = SummaryWriter()  # 初始化 TensorBoard，用于可视化损失变化


    # 训练参数
    n_epochs = config['epochs']
    best_loss = math.inf  # 初始化最佳验证损失
    step = 0  # 训练步数
    early_stop_count = 0  # 早停计数器

    # 训练循环
    for epoch in range(n_epochs):
        model.train()  # 设置模型为训练模式
        loss_list = []  # 存储训练损失
        train_pbar = tqdm(train_loader, position=0, leave=True)  # tqdm 进度条

        for x, y in train_pbar:
            optimizer.zero_grad()  # 清空梯度
            x, y = x.to(device), y.to(device)  # 数据迁移到 GPU 或 CPU

            pred = model(x)  # 前向传播
            loss = criterion(pred, y.unsqueeze(-1))  # 计算损失，y.unsqueeze(-1) 保持维度一致
            loss.backward()  # 反向传播计算梯度
            optimizer.step()  # 更新参数

            step += 1
            loss_list.append(loss.detach().item())  # 记录 loss

            train_pbar.set_description(f'Epoch [{epoch+1}/{n_epochs}]')
            train_pbar.set_postfix({'loss': loss.detach().item()})

        # 计算当前 epoch 的平均训练损失
        mean_train_loss = sum(loss_list) / len(loss_list)  # ✅ 计算平均损失

        writer.add_scalar('Loss/train', mean_train_loss, step)  # 记录训练损失

        # **验证阶段**
        model.eval()  # 进入验证模式
        loss_record = []  # 存储验证损失
        with torch.no_grad():
            for x, y in val_loader:
                x, y = x.to(device), y.to(device)
                pred = model(x)
                loss = criterion(pred, y.unsqueeze(-1))
                loss_record.append(loss.item())

        mean_valid_loss = sum(loss_record) / len(loss_record) if len(loss_record) > 0 else 0

        print(f'Epoch [{epoch+1}/{n_epochs}]: Train loss: {mean_train_loss:.4f}, Valid loss: {mean_valid_loss:.4f}')
        writer.add_scalar('Loss/valid', mean_valid_loss, step)  # 记录验证损失

        # **保存最优模型**
        if mean_valid_loss < best_loss:
            best_loss = mean_valid_loss
            torch.save(model.state_dict(), config['save_path'])  # 保存最佳模型
            print(f'Saving model with loss {best_loss:.4f}...')
            early_stop_count = 0  # 重置早停计数器
        else:
            early_stop_count += 1  # 早停计数 +1

        # **早停机制**
        if early_stop_count >= config['early_stop']:
            print('\nModel is not improving, stopping training.')
            break  # 提前终止训练

        # **动态调整学习率**
        scheduler.step()  
        current_lr = optimizer.param_groups[0]['lr']  # 获取当前学习率
        print(f'Learning Rate: {current_lr:.12f}')  # 输出当前学习率
        writer.add_scalar('Learning Rate', current_lr, step)  # 记录学习率变化

    print('Training Complete!')
```

```python
model = LSTM(X_train_tensor.shape[2]).to(device) # put your model and data on the same computation device.
trainer(train_loader, val_loader, model, config, device)
```

```python
# 加载模型
model = LSTM(input_size=X_train_tensor.shape[2])  # 重新创建模型
model.load_state_dict(torch.load("./models/best_model.pth"))  # 加载参数

# 让模型进入评估模式
model.eval()
print("✅ Model loaded and set to evaluation mode.")
# 进行预测（使用 `with torch.no_grad()` 以节省显存）
#评估模型
def evaluate_model(model, test_loader):
    model.eval()
    predictions,actuals = [],[]
    with torch.no_grad():
        for i,(x_batch,y_batch) in enumerate(test_loader):
            x_batch = x_batch.to(device)
            y_batch = y_batch.to(device)

            outputs = model(x_batch)
            predictions.extend(outputs.cpu().numpy())
            actuals.extend(y_batch.cpu().numpy())

    return np.array(predictions),np.array(actuals)





```
